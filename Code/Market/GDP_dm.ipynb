```python
# Get data from index 0 to 216 and row 265
data= pd.concat([df.iloc[0:217], df.iloc[[265]]])

# Display the data
data
```

```python
# Define a dictionary of patterns and their replacements for various indicators across years
replacement_patterns = {
    # Agriculture
    r'(\d{4}) \[YR\1\] - Agriculture, forestry, and fishing, value added \(% of GDP\) \[NV\.AGR\.TOTL\.ZS\]': r'\1- Agriculture (% of GDP)',
    r'(\d{4}) \[YR\1\] - Agriculture, forestry, and fishing, value added \(current US\$\) \[NV\.AGR\.TOTL\.CD\]': r'\1- Agriculture (current US$)',

    # Industry
    r'(\d{4}) \[YR\1\] - Industry \(including construction\), value added \(current US\$\) \[NV\.IND\.TOTL\.ZS\]': r'\1- Industry (% of GDP)',
    r'(\d{4}) \[YR\1\] - Industry \(including construction\), value added \(current US\$\) \[NV\.IND\.TOTL\.CD\]': r'\1- Industry (current US$)',

    # Services
    r'(\d{4}) \[YR\1\] - Services, value added \(% of GDP\) \[NV\.SRV\.TOTL\.ZS\]': r'\1- Services (% of GDP)',
    r'(\d{4}) \[YR\1\] - Services, value added \(current US\$\) \[NV\.SRV\.TOTL\.CD\]': r'\1- Services (current US$)',

    # GDP
    r'(\d{4}) \[YR\1\] - GDP \(current US\$\) \[NY\.GDP\.MKTP\.CD\]': r'\1- GDP (current US$)',
    r'(\d{4}) \[YR\1\] - GDP growth \(annual %\) \[NY\.GDP\.MKTP\.KD\.ZG\]': r'\1- GDP growth (annual %)',

    # Inflation
    r'(\d{4}) \[YR\1\] - Inflation, consumer prices \(annual %\) \[FP\.CPI\.TOTL\.ZG\]': r'\1- Inflation (annual %)',

}

# Loop through the patterns and apply the replacements to all columns
for pattern, replacement in replacement_patterns.items():
    data.columns = data.columns.str.replace(pattern, replacement, regex=True)

# Print the updated column names to verify the changes
#print(data.columns)
data
```

```python
# Update the pattern specifically for this column
data.columns = data.columns.str.replace(
    r'(\d{4}) \[YR\1\] - Industry \(including construction\), value added \(% of GDP\) \[NV\.IND\.TOTL\.ZS\]', 
    r'\1- Industry (% of GDP)', 
    regex=True
)

# Print the updated column names to verify the changes
#print(data.columns)
data
```

```python
data.dtypes
```

```python
# Convert all columns except 'Country Name' and 'Country Code' to numeric, coercing errors
numeric_columns = data.columns.difference(['Country Name', 'Country Code'])

# Apply pd.to_numeric to all these columns
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
data.dtypes
```

```python
# using simple imputer from scikit-learn to handle the missing values
from sklearn.impute import SimpleImputer

categorical_cols = data.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = data.select_dtypes(include=["number"]).columns.tolist()

# Replacing the null values in numerical columns with mean
#imputer_numerical = SimpleImputer(strategy='mean')
# Forward fill using the more explicit method
data[numerical_cols] = data[numerical_cols].ffill().bfill()

# Replacing the null values in categorical columns with most-frequent item in the feature
imputer_categorical = SimpleImputer(strategy='most_frequent')
data[categorical_cols] = imputer_categorical.fit_transform(data[categorical_cols])
```

```python
#data rounding off
data[numerical_cols] = data[numerical_cols].round(3)
pd.options.display.float_format = '{:.3f}'.format
columns_to_convert = [col for col in data.columns if 'US$' in col]
data[columns_to_convert] = data[columns_to_convert].apply(lambda x: x / 1e9)

# Update the column names to reflect "US$ in Billions"
data.rename(columns={col: col.replace('current US$', 'US$ in Billions') for col in columns_to_convert}, inplace=True)
```

### Data Visualization

```python

# Filter the data for the U.S. only
us_data = data[data['Country Name'] == 'United States']

# Extract the years and corresponding sectoral contributions to GDP
years = list(range(2013, 2024))
agriculture_us = us_data[[f'{year}- Agriculture (% of GDP)' for year in years]].values.flatten()
industry_us = us_data[[f'{year}- Industry (% of GDP)' for year in years]].values.flatten()
services_us = us_data[[f'{year}- Services (% of GDP)' for year in years]].values.flatten()

# Extract the U.S. GDP and World GDP over the years
us_gdp = us_data[[f'{year}- GDP (US$ in Billions)' for year in years]].values.flatten()

# Filter the data for World (assuming "World" is available as a country in the dataset)
world_data = data[data['Country Name'] == 'World']
world_gdp = world_data[[f'{year}- GDP (US$ in Billions)' for year in years]].values.flatten()

# Create the trend line plot for sectors with an additional line for U.S. and World GDP
plt.figure(figsize=(10, 6))

# Sectoral contributions for U.S.
plt.plot(years, agriculture_us, label='U.S. Agriculture (% of GDP)', marker='o')
plt.plot(years, industry_us, label='U.S. Industry (% of GDP)', marker='o')
plt.plot(years, services_us, label='U.S. Services (% of GDP)', marker='o')

# Add U.S. and World GDP (scale them for comparison)
plt.plot(years, us_gdp / 1000, label='U.S. GDP (Trillions US$)', color='black', linestyle='--', marker='x')
plt.plot(years, world_gdp / 10000, label='World GDP (Trillions US$)', color='red', linestyle='--', marker='x')

# Customize the plot
plt.title('U.S. Sector-wise Contributions and GDP vs World GDP (2013-2023)')
plt.xlabel('Year')
plt.ylabel('Percentage / Trillions of US$')
plt.xticks(years)
plt.grid(True)
plt.legend()
plt.savefig("U.S. Sector-wise Contributions and GDP vs World GDP (2013-2023).jpg", dpi=300)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# List of G20 countries (Country names as they might appear in the dataset)
g20_countries = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 
                 'Indonesia', 'Italy', 'Japan', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 
                 'South Korea', 'Turkey', 'United Kingdom', 'United States', 'European Union']

# Clean the column names by stripping any leading or trailing spaces
data.columns = data.columns.str.strip()

# Filter the dataset for G20 countries
g20_data = data[data['Country Name'].isin(g20_countries)]

# Extract the years
years = list(range(2013, 2024))

# Extract the GDP growth data for G20 countries for each year from 2013 to 2023
gdp_growth_g20 = g20_data[[f'{year}- GDP growth (annual %)' for year in years]].mean()

# Create a line plot for G20 GDP growth from 2013 to 2023
plt.figure(figsize=(10, 6))

plt.plot(years, gdp_growth_g20, marker='o', color='blue', label='G20 GDP Growth (annual %)')

# Customize the plot
plt.title('Average GDP Growth for G20 Countries (2013-2023)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.xticks(years)
plt.grid(True)
plt.legend()
plt.savefig("Average GDP Growth for G20 Countries (2013-2023).jpg", dpi=300)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Extract inflation rates for G20 countries in 2023
inflation_2023_g20 = g20_data[['Country Name', '2023- Inflation (annual %)']].dropna()

# Create a bar plot for G20 countries' inflation rates in 2023
plt.figure(figsize=(12, 6))

plt.bar(inflation_2023_g20['Country Name'], inflation_2023_g20['2023- Inflation (annual %)'], color='teal')

# Customize the plot
plt.title('Inflation Rates for G20 Countries (2023)')
plt.xlabel('Country')
plt.ylabel('Inflation Rate (annual %)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.savefig("Inflation Rates for G20 Countries (2023)'.jpg", dpi=300)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

### Step 1: Label Encoding the Categorical Columns

# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'Country Name' column
data['Country Name Encoded'] = label_encoder.fit_transform(data['Country Name'])

### Step 2: Standardizing the Numerical Columns

# Define the numerical columns you want to standardize from 2018 to 2023
numerical_columns = [
    col for col in data.columns if 
    ('2018-' in col or '2019-' in col or '2020-' in col or '2021-' in col or '2022-' in col or '2023-' in col) and 
    ('GDP (US$ in Billions)' in col or 'Inflation (annual %)' in col or 'GDP growth (annual %)' in col or
     'Agriculture (% of GDP)' in col or 'Industry (% of GDP)' in col or 'Services (% of GDP)' in col)
]

# Drop rows with missing data in the selected numerical columns
data_cleaned = data.dropna(subset=numerical_columns)

# Apply StandardScaler to the numerical columns
scaler = StandardScaler()
standardized_numerical = scaler.fit_transform(data_cleaned[numerical_columns])

# Convert the standardized data back into a DataFrame with appropriate column names
standardized_numerical_df = pd.DataFrame(standardized_numerical, columns=numerical_columns)

### Step 3: Concatenating the Encoded and Standardized Data

# Concatenate the label encoded 'Country Name Encoded' column with the standardized numerical data
final_data = pd.concat([data['Country Name'].reset_index(drop=True), 
                        data['Country Name Encoded'].reset_index(drop=True), 
                        standardized_numerical_df.reset_index(drop=True)], axis=1)

# Display the final dataset
final_data
```
