```python
import pandas as pd
```

```python
citydata = pd.read_csv("WeatherDataSet1/city_info.csv")
citydata.head()
```

```python
list_of_cities = citydata["Name"].unique().tolist()
```

```python
list_of_cities
```

```python
list_of_codes=[]
for city in list_of_cities:
    result = citydata.loc[citydata['Name']==city, 'ID']
    list_of_codes.append(result.iloc[0])

list_of_codes
```

```python
import pandas as pd
import os
directory =os.path.join(os.getcwd(),"WeatherDataSet1")

for i in range(37,len(list_of_codes)):
    df= pd.read_csv(f"WeatherDataSet1/{list_of_codes[i]}.csv")
    if os.path.exists("WeatherDataSet1/combined_dataset.csv"):
        main_data = pd.read_csv("WeatherDataSet1/combined_dataset.csv")
        main_data = pd.concat([main_data, df], ignore_index = True)
    else:
        main_data = pd.DataFrame()
        main_data = pd.concat([main_data, df], ignore_index=True)
    main_data.to_csv("WeatherDataSet1/combined_dataset.csv",index=False)
    print(f"added {list_of_cities[i]} to main")
```
